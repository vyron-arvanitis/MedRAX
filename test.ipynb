```python
tool_outputs=[
    {"tool":"chest_xray_classifier","prediction":"cardiomegaly","confidence":0.78,"prior":0.88},
    {"tool":"chest_xray_classifier2","prediction":"cardiomegaly","confidence":0.88,"prior":0.88},
    {"tool":"chest_xray_classifier3","prediction":"cardiomegaly","confidence":0.58,"prior":0.28}

]

scores = []
for out in tool_outputs:
    conf = out["confidence"]
    prior = out["prior"]
    scores.append({"tool" : out["tool"], "score": conf*prior})

scores.sort(key=lambda x: x["score"], reverse=True)
top =scores[0]["score"]
top
```
